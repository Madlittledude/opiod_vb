```python




def PDF2Text():
    import os
    import pytesseract
    from pdf2image import convert_from_path
    from pdf2image.exceptions import PDFPageCountError
    import time

    starttime = time.time()
    # Set the path to your files
    pathtoPDF = 'Library'
    pathtoText = 'LIB_TEXT'
    pagelimit = 1000
    doc = 0
    times = []
    files_processed = 0
    for root, dirs, files in os.walk(pathtoPDF):
        if not files:
            print("There are no files in the path.")
            return

        for f in files:
            # Set the filename of your PDF file

            # Set the resolution of the converted images
            resolution = 200
            print(f)

            endtime = time.time()
            elapsetime = endtime - starttime
            times.append(elapsetime)
            path = os.path.join(root, f)
            if '.' in f[:3]:
                continue
            print(f)
            print(f"Time Elapsed: {elapsetime:.2f} seconds")
            doc += 1
            # Convert the PDF to a list of image pages
            pages = convert_from_path(path, resolution)
            # Loop through each page and perform OCR on the image
            for i, page in enumerate(pages):

                if i == pagelimit:  # page limit
                    break
                try:
                    # Convert the image page to grayscale
                    image = page.convert('L')

                    # Use Tesseract to perform OCR on the image and print the text
                    text = pytesseract.image_to_string(image)

                    subdir = root.split(pathtoPDF)[-1]
                    outputfile = os.path.join(pathtoText, subdir, f.split('.')[0])
                    # Create the output directory if it doesn't exist
                    os.makedirs(os.path.dirname(outputfile), exist_ok=True)
                    # Write the OCR text to the text file
                    if i == 0:
                        with open(outputfile + '.txt', 'w') as file:
                            file.write(text)
                            print(f'{f}, Page: {i + 1}')
                            print(f'{doc}/{len(files)}')
                    else:
                        with open(outputfile + '.txt', 'a') as file:
                            file.write(text)
                            print(f'{f}, Page: {i + 1}')
                            print(f'{doc}/{len(files)}')
                except PDFPageCountError:
                    print(f'Error: Unable to extract page {i + 1} from {f}')
                    continue

            files_processed += 1


            # clean_text('TEXT/'+f.split('.')[0]+'.txt')
        endtime = time.time()
        if files_processed > 0:
            print(elapsetime, sum(times) // files_processed)
        else:
            print("There are no files in the path.")
        
    
    # rename()

PDF2Text()

```

```python
from llama_index import SimpleDirectoryReader
filename_fn = lambda filename: {'file_name': filename}

# automatically sets the metadata of each document according to filename_fn
documents = SimpleDirectoryReader('LIB_TEXT', file_metadata=filename_fn).load_data()
```

```python
from llama_index.node_parser import SimpleNodeParser

parser = SimpleNodeParser()

nodes = parser.get_nodes_from_documents(documents)
```

```python
from llama_index import VectorStoreIndex
# construct the index with the txt document
index = VectorStoreIndex.from_documents(documents)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("What are some examples of a drug that can be used to treat opioid addiction? and what're their distinctions?")
```
